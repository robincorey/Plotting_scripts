```python
import sys
import numpy as np
np.set_printoptions(threshold=sys.maxsize)
import matplotlib
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
```

```python
def getdata (file):
        x, y = np.loadtxt(fname=file, usecols=(0,1), comments=['#','@'], unpack=True)
        return x, y
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
hfont = {'fontname':'Helvetica'}


x, y = getdata('2Dproj_noligand.1.A.xvg') #2D_noligand.all.xvg')
xy = np.vstack([x,y])
z = gaussian_kde(xy)(xy)
im = ax.scatter(x,y, c=z, s=5)

ax.set_xlim(-12,10)
ax.set_xticks(np.arange(-10,11,step=5), fontsize=30)
ax.set_xlabel('xlabel', **hfont)
ax.set_ylim(-10,10)
ax.set_yticks(np.arange(-10,11,step=5))
ax.yaxis.set_tick_params(width=2, length=8)
for axis in ['bottom','left']:
    ax.spines[axis].set_linewidth(2)
right_side = ax.spines["right"]
right_side.set_visible(False)
top = ax.spines["top"]
top.set_visible(False)

plt.colorbar(im)
#plt.show()
```
