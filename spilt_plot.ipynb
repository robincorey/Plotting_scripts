```python
import numpy as np
import matplotlib.pyplot as plt
#from matplotlib import transforms
```

```python
def plotdata (fname):
    fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,5), gridspec_kw={'width_ratios': [8, 1]})
    x, y = np.loadtxt(fname,usecols=[0,1],comments=['#','@'], unpack=True)
    ax1.plot(x/1000,y,color='red')
    ax1.set(xlabel='time (ns)', ylabel='distance (nm)')
    #base = plt.gca().transData
    #rot = transforms.Affine2D().rotate_deg(90)
    ax2.hist(y, orientation=u'horizontal')
    ax2.set(xlabel='density')
```

```python
def plotdatashare (fname):
    fig = plt.figure()
    #fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,5), gridspec_kw={'width_ratios': [8, 1]})
    gs = fig.add_gridspec(1,2, wspace=0, width_ratios=[8,1])
    axs = gs.subplots(sharey=True)
    x, y = np.loadtxt(fname,usecols=[0,1],comments=['#','@'], unpack=True)
    axs[0].plot(x/1000,y,color='red')
    axs[0].set(xlabel='time (ns)', ylabel='distance (nm)')
    #base = plt.gca().transData
    #rot = transforms.Affine2D().rotate_deg(90)
    axs[1].hist(y, orientation=u'horizontal', color='gray')
    axs[1].set(xlabel='density')
```

```python
plotdatashare('data.xvg')
```
