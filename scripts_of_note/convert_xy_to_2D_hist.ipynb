```python
import sys
import numpy as np
np.set_printoptions(threshold=sys.maxsize)
import matplotlib
import matplotlib.pyplot as plt
from scipy.interpolate import interpn
```

```python
def getdata (file):
        x, y = np.loadtxt(fname=file, usecols=(0,1), comments=['#','@'], unpack=True)
        return x, y
```

```python
# this is a script which allows 2D xy data (e.g. from PCA 2D proj) to be combined into a 2D histogram
# and then plotted, or the output given for plotting in a different system e.g. Prism
# note that the printed output needs to be reformatted a bit (vim/sed)

x, y = getdata('2D_noligand.all.xvg') 
data  = plt.hist2d(x, y, bins=np.arange(-10.5,11.5,step=0.5), cmap=plt.cm.jet)
plt.show()
print(data[1]) # x axis
print(data[2]) # y axis
print(data[0]) # data
```
