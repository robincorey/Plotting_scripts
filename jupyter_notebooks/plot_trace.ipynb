```python
import numpy as np
import matplotlib.pyplot as plt
#from matplotlib import transforms
```

```python
def plotdata (fname, col, title, i):
    #fig, ax = plt.subplots(1, figsize=(10,5))
    x, y = np.loadtxt(fname,usecols=[0,3],comments=['#','@'], unpack=True)
    plt.plot(x/1000,-y,color=col, label='run %s' % i)
    #plt.ylim(-1,1.5)
    plt.title(title, fontsize=15)
    plt.xlabel('time (ns)', fontsize=15)
    plt.legend(frameon=False)
    plt.ylabel('distance (nm)', fontsize=15)
```

```python
cols = [ 'black', 'red', 'blue', 'orange']
```

```python
for i in np.arange(1,4):
    plotdata('PNTM.1V.%s.xvg' % i, cols[i], '+ve potential', i)
```

```python
for i in np.arange(1,4):
    plotdata('PNTM.minus1V.%s.xvg' % i, cols[i], '-ve potential', i)
```
